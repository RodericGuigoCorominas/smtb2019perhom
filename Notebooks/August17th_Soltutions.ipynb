# First Homology: Holes

```python
ctop=[(1.-(float(x)/100)**2)**0.5 for x in range(-100,100)]
cbot=[-(1.-(float(x)/100)**2)**0.5 for x in range(-100,100)]
circper=np.array([[float(i)/100,ctop[i+100]] for i in range(-100,100)]+[[float(i)/100,cbot[i+100]] for i in range(-100,100)])
```

```python
plt.plot(circper[:,0],circper[:,1],"o")
plt.show()
```

```python
diagcirc = ripser(circper)['dgms']
plot_diagrams(diagcirc, show=True)
```

```python
def sample_spherical(npoints, ndim=3):
    vec = np.random.randn(npoints,ndim)
    vec = [v/np.linalg.norm(v, axis=0) for v in vec]
    return vec

sphere=sample_spherical(100)
xi = [pt[0] for pt in sphere]
yi = [pt[1] for pt in sphere]
zi = [pt[2] for pt in sphere] 

fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})
#ax.plot_wireframe(x, y, z, color='k', rstride=1, cstride=1)
ax.scatter(xi, yi, zi, s=100, c='r', zorder=10)
```

```python
ph_sphere=ripser(np.array(sphere),maxdim=2)['dgms']
plot_diagrams(ph_sphere, show=True)
```

```python
sphere_noise_1=np.array([pt+[c for c in sample_spherical(1)[0]] for pt in sphere])
xin = [pt[0] for pt in sphere_noise_1]
yin = [pt[1] for pt in sphere_noise_1]
zin = [pt[2] for pt in sphere_noise_1] 
fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})
#ax.plot_wireframe(x, y, z, color='k', rstride=1, cstride=1)
ax.scatter(xin, yin, zin, s=100, c='r', zorder=10)
```

```python
ph_sphere_1=ripser(sphere_noise_1,maxdim=2)['dgms']
plot_diagrams(ph_sphere_1, show=True)
```

```python
sphere_noise_3=np.array([pt+[c/3 for c in sample_spherical(1)[0]] for pt in sphere])
xin = [pt[0] for pt in sphere_noise_3]
yin = [pt[1] for pt in sphere_noise_3]
zin = [pt[2] for pt in sphere_noise_3] 
fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'equal'})
#ax.plot_wireframe(x, y, z, color='k', rstride=1, cstride=1)
ax.scatter(xin, yin, zin, s=100, c='r', zorder=10)
```

```python
ph_sphere_3=ripser(sphere_noise_3,maxdim=2)['dgms']
plot_diagrams(ph_sphere_3, show=True)
```
